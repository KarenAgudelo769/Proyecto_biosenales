```python
import mne
import os
import glob
import numpy as np
import pandas as pd
from tabulate import tabulate

# --- DEFINICIÓN DE FUNCIONES DE PROCESAMIENTO ---

def procesar_archivo(ruta):
    """
    Carga el archivo .set, aplica filtros, selecciona canales y extrae épocas.
    """
    # Extraer metadatos desde el nombre del archivo
    nombre = os.path.basename(ruta)
    partes = nombre.split('_')
    sujeto = partes[0]
    run = partes[2].split('-')[1]
    
    # Clasificar condición según el número de run (PhysioNet)
    if run in ['4', '8', '12']:
        condicion = "imaginacion"
    elif run in ['3', '7', '11']:
        condicion = "real"
    else:
        condicion = "otra"
    
    # Cargar datos EEG pre-cargados en memoria
    raw = mne.io.read_raw_eeglab(ruta, preload=True, verbose=False)
    
    # Selección de canales de interés (Corteza Motora)
    raw.pick(['C3', 'Cz', 'C4'])
    
    # Aplicar filtro Notch (60Hz) y filtro de banda (0.5 - 45 Hz)
    raw.notch_filter(60, verbose=False)
    raw.filter(0.5, 45, verbose=False)
    
    # Extraer eventos y mapear IDs de tareas
    events, event_id = mne.events_from_annotations(raw, verbose=False)
    
    # Definir diccionario de eventos para segmentación
    eventos_dict = {
        'reposo':    event_id.get('TASK1T0') or event_id.get('TASK2T0'),
        'izquierda': event_id.get('TASK1T1') or event_id.get('TASK2T1'),
        'derecha':   event_id.get('TASK1T2') or event_id.get('TASK2T2')
    }
    
    # Filtrar eventos nulos no presentes en el archivo
    eventos_dict = {k: v for k, v in eventos_dict.items() if v is not None}
    
    if not eventos_dict:
        return None, sujeto, run, condicion

    # Segmentar señal en épocas de 2 segundos
    epocas = mne.Epochs(
        raw, events, event_id=eventos_dict,
        tmin=0, tmax=2,
        baseline=None,
        preload=True,
        verbose=False
    )
    
    return epocas, sujeto, run, condicion
```

```python
def calcular_psd_epochs(epocas):
    """
    Calcula la Densidad Espectral de Potencia (PSD) mediante el método de Welch.
    """
    psds = epocas.compute_psd(
        method='welch', fmin=0, fmax=45, 
        n_fft=256, n_overlap=128, verbose=False
    )
    return psds.get_data(), psds.freqs

def potencia_banda(psds, freqs, fmin, fmax):
    """
    Calcula el promedio de potencia en un rango de frecuencias determinado.
    """
    idx = (freqs >= fmin) & (freqs <= fmax)
    return np.mean(psds[:, :, idx], axis=2)
```

```python
def construir_dataframe(epocas, sujeto, run, condicion):
    """
    Transforma los datos de potencia de épocas en un DataFrame estructurado.
    """
    filas = []
    canales = epocas.ch_names
    
    # Procesar potencias Mu y Beta por cada tarea disponible
    for tarea in ['reposo', 'izquierda', 'derecha']:
        if tarea not in epocas.event_id:
            continue
        
        ep = epocas[tarea]
        psds, freqs = calcular_psd_epochs(ep)
        
        # Extraer potencia media en bandas Mu (8-13Hz) y Beta (13-30Hz)
        mu = potencia_banda(psds, freqs, 8, 13)
        beta = potencia_banda(psds, freqs, 13, 30)
        
        n_epochs = mu.shape[0]
        
        # Almacenar resultados por época y canal en la lista de filas
        for i in range(n_epochs):
            for ch_idx, canal in enumerate(canales):
                filas.append({
                    'sujeto': sujeto,
                    'run': run,
                    'condicion': condicion, 
                    'tarea': tarea,
                    'canal': canal,
                    'mu': mu[i, ch_idx],
                    'beta': beta[i, ch_idx]
                })
    
    return pd.DataFrame(filas)
```

```python
# --- EJECUCIÓN DEL FLUJO DE TRABAJO ---

# Definir directorio de origen y listar archivos .set
carpeta_sujetos = 'sujetos'
archivos = glob.glob(os.path.join(carpeta_sujetos, "*.set"))
lista_dfs = []

print(f"Archivos encontrados para procesar: {len(archivos)}")

# Ejecutar bucle de procesamiento sobre la lista de archivos
for ruta in archivos:
    try:
        # Procesar señales y extraer épocas
        epocas, sujeto, run, condicion = procesar_archivo(ruta)
        
        if epocas is not None:
            # Construir DataFrame temporal y añadir a la lista global
            df_temp = construir_dataframe(epocas, sujeto, run, condicion)
            lista_dfs.append(df_temp)
            print(f"Procesado exitosamente: {os.path.basename(ruta)}")
            
    except Exception as e:
        print(f"Error en el archivo {ruta}: {e}")

# Consolidar todos los DataFrames si la lista contiene elementos
if lista_dfs:
    # Concatenar resultados en el objeto df_final
    df_final = pd.concat(lista_dfs, ignore_index=True)
    
    # Mostrar resumen de datos procesados
    print("\n" + "="*60)
    print("REPORTE DE DATOS CONSOLIDADOS")
    print("="*60)
    
    # Imprimir primeras 50 filas con formato de tabla
    print(tabulate(df_final.head(50).round(6), headers='keys', tablefmt='grid', showindex=False))
    print(f"\nTotal de registros procesados: {len(df_final)}")
else:
    print("No se generaron datos. Verifique el contenido de la carpeta 'sujetos'.")
```
